```python
import os
import pandas as pd
```

```python
gt_file = r'C:\Users\z004b0je\OneDrive - Siemens Healthineers\Projects\20_Scientific_Proof_Points\shoulder_mr\gt_result\GT_Xie_Tao.xlsx'
ai_file = r'C:\Users\z004b0je\OneDrive - Siemens Healthineers\Projects\20_Scientific_Proof_Points\shoulder_mr\results_tables\ai_data_v1.xlsx'
```

```python
df_gt = pd.read_excel(gt_file)
df_ai = pd.read_excel(ai_file)
```

```python
df_gt.head(3)
```

```python
df_ai.head(3)
```

```python
result_info = dict()
cols = ['PID', 'GT_Supraspinatus', 'GT_Infraspinatus', 'GT_Subscapularis', 'AI_Supraspinatus', 'AI_Infraspinatus', 'AI_Subscapularis']
for col in cols:
    result_info.setdefault(col, [])

for index,row in df_gt.iterrows():
    result_info['PID'].append(row.PID)
    result_info['GT_Supraspinatus'].append(row.GT_Supraspinatus)
    result_info['GT_Infraspinatus'].append(row.GT_Infraspinatus)
    result_info['GT_Subscapularis'].append(row.GT_Subscapularis)
    if row.PID in df_ai.PID.values:
        result_info['AI_Supraspinatus'].append(df_ai[df_ai['PID'] == row.PID].Supraspinatus_AI.values[0])
        result_info['AI_Infraspinatus'].append(df_ai[df_ai['PID'] == row.PID].Infraspinatus_AI.values[0])
        result_info['AI_Subscapularis'].append(df_ai[df_ai['PID'] == row.PID].Subscapularis_AI.values[0])
    else:
        result_info['AI_Supraspinatus'].append('NA')
        result_info['AI_Infraspinatus'].append('NA')
        result_info['AI_Subscapularis'].append('NA')

result_df = pd.DataFrame(result_info)
result_df.to_excel(os.path.join(r'C:\Users\z004b0je\OneDrive - Siemens Healthineers\Projects\20_Scientific_Proof_Points\shoulder_mr\results_tables','GT_AI.xlsx'), index=False)
    
    
```
