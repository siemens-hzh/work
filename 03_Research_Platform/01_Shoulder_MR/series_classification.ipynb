```python
import os
import pandas as pd
import pydicom
from tqdm import tqdm
from sklearn.linear_model import LogisticRegression
```

```python
right_folder = r'F:\huashan\correct\right'
left_folder = r'F:\huashan\correct\left'
```

```python
right_ax, right_sag, right_cor,right_no_fs = [],[],[],[]


for case in os.listdir(right_folder):
    case_folder = os.path.join(right_folder, case)
    series_folders = os.listdir(case_folder)
    for series_folder in series_folders:
        cur_folder = os.path.join(case_folder, series_folder)
        ds = pydicom.dcmread(os.path.join(cur_folder, os.listdir(cur_folder)[0]))
        orientation = ds.ImageOrientationPatient
        image_orientation = [round(x, 2) for x in orientation]
        #print(image_orientation)
        series_words = series_folder.split('_')
        if 'fs' in series_words or 'stir' in series_words or 'fat' in series_words or 'spair' in series_words:
            if 'cor' in series_words or 'co' in series_words or 'ocor' in series_words:
                image_orientation.append(3)
                right_cor.append(image_orientation)
            elif 'sag' in series_words or 'osag' in series_words or 'sg' in series_words:
                image_orientation.append(4)
                right_sag.append(image_orientation)
            elif 'ax' in series_words or 'oax' in series_words or 'tra' in series_words or 'axi' in series_words:
                image_orientation.append(0)
                right_ax.append(image_orientation)
        else:
            if 'cor' in series_words or 'ocor' in series_words or 'co' in series_words:
                right_no_fs.append(image_orientation)
                    
print(right_cor)

            
```
