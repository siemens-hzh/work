```python
import os
import pandas as pd

input_file = r'C:\Users\z004b0je\OneDrive - Siemens Healthineers\Projects\20_Scientific_Proof_Points\huashan_LLM\pancreas_reports.xlsx'
df = pd.read_excel(input_file)
df.head(5)
```

```python
new_findings = []
check_cols = ['肠系膜上动脉','腹腔干','肝总动脉','腹主动脉','动脉变异','门静脉','肠系膜上静脉','下腔静脉']
for index, row in df.iterrows():
    r_findings = row.放射学表现
    try:
        for col in check_cols:
            if row[col] ==0:
                continue
            elif row[col] == 1:
                r_findings += f'占位与{col}的接触小于180度。'
            elif row[col] == 2:
                r_findings += f'占位与{col}的接触大于180度。' 
            else:
                
                r_findings += str(row[col])            
        r_findings += '占位与其他血管无接触。'
        
    except Exception as e:
        print(index, e)
        continue
    new_findings.append(r_findings)
df['new_findings'] = new_findings
```

```python
df.to_excel(r'C:\Users\z004b0je\OneDrive - Siemens Healthineers\Projects\20_Scientific_Proof_Points\huashan_LLM\input_reports.xlsx', index=False)
```
